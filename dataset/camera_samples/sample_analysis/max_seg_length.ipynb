```python
import numpy as np
from pathlib import Path
```

```python
# Dataset directory
dataset_dir = Path("../water_bottle_dataset3")

# Number of views to check 
num_views = 1000

# Initialize variables to track the largest segmentation
max_length = 0
max_view_id =  None
max_shape = None
```

```python
# Iterate through all views
for i in range(num_views):
    view_id = f"view_{i}"  
    view_dir = dataset_dir / view_id
    
    if not view_dir.exists():
        print(f"Warning: {view_dir} does not exist. Skipping.")
        continue
    
    # Load segmentation array
    seg_path = view_dir / "segmentation.npy"
    if not seg_path.exists():
        print(f"Warning: {seg_path} does not exist. Skipping.")
        continue
    
    seg = np.load(seg_path)
    bottle_pixels = np.where(seg == 1)[0]  # Indices where (seg == 1) indicates bottle presence
    length = len(bottle_pixels)
    
    print(f"View {view_id}: length={length}, shape={seg.shape}")
    
    # Update maximum if current length is larger
    if length > max_length:
        max_length = length
        max_view_id = view_id
        max_shape = seg.shape

```

```python
# Print results
if max_view_id is not None:
    print(f"\nExtremes segmentation array:")
    print(f"Max view: {max_view_id}")
    print(f"Max length: {max_length} (number of bottle pixels)")
    print(f"Max shape: {max_shape}")

else:
    print("\nNo segmentation arrays found.")
```
